```python
### Read in your test data and create the arrays X (features) and y (labels) to train the neural network

import numpy as np
import keras
import tensorflow as tf
from keras_preprocessing import sequence
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle
from scipy import ndimage
from PIL import Image
import imageio

DATADIR = "Datasets"
CATEGORIES = ["Pascal", "Stephanie", "Claudia", "Daniel", "David", "Emely", "Eric", "Hannah", "Heike", "Merlin", "Nicole", "Sascha"]

training_data = []
datasize1 = 60
datasize2 = 20

def create_training_data():

    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)
        index_num = CATEGORIES.index(category)
        for dataset in os.listdir(path):
            try:
                pickle_path = os.path.join(path,dataset) 
                pickle_file = pickle.load(open(pickle_path,"rb")) # Liest Datensatz ein
                np_array = np.array(pickle_file) # Wandelt Datensatz in ein Numpy-Array um
                #print(np_array.shape)
                reshaped_array = np_array.reshape(-1, np_array.shape[1]) # Wandelt das 3D-Array in 2D-Array um
                #print(reshaped_array)
                final_array = np.resize(reshaped_array, (datasize1,datasize2)) # Bringt alle Datens채tze(Arrays) in die selbe Form - Evtl anpassen, wenn sich Anzahl der Samples ver채ndern?
                #print(final_array.shape)
                training_data.append([final_array, index_num])
            except Exception as e:
                pass
    #print(training_data)         


create_training_data()   

random.shuffle(training_data)   

X = []
y = []

for features, label in training_data: # Features werden in X, labels in y gespeichert
    X.append(features)
    y.append(label)
X = np.array(X) 
X = X.reshape(list(X.shape) + [1]) ## Bringt das Array in eine 4-Dimensionale Form. Notwendig f체r das neuronale Netzwert

print("Shape of X:")
print(X.shape)           
y = np.array(y)



print("Shape of y:")
print(y.shape)

pickle_out = open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

with open("y.pickle", "wb") as f:
    pickle.dump(y, f)
```
